```python
list_variales = []
class_var = []
c=1
for folder in os.listdir():
    for images in os.listdir(os.getcwd()+"\\{}".format(folder)):
        img = cv2.imread(os.getcwd()+"\\{}\\{}".format(folder,images),0)
        img = cv2.resize(img, (50,50))
        list_variales.append(img.flatten())
        class_var.append(folder)
        c+=1
```
