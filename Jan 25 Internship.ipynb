```python
import json
# Load the dataset
df = pd.read_csv(r"C:\Users\GOLLA RAVI TEJA\Downloads\property_photos.tsv", sep='\t')

# Parse the 'photo_urls' column
def parse_photo_urls(photo_urls):
    try:
        # Convert the string to a list of dictionaries
        photo_list = json.loads(photo_urls)
        # Extract specific fields, e.g., titles and names
        titles = [photo['title'] for photo in photo_list]
        names = [photo['name'] for photo in photo_list]
        return titles, names
    except (TypeError, json.JSONDecodeError):
        # Handle NaN or invalid entries
        return None, None

# Apply parsing and create new columns
df[['titles', 'names']] = df['photo_urls'].apply(lambda x: pd.Series(parse_photo_urls(x)))

# Preview the updated dataframe
print(df.head())
```

```python
import os
import pandas as pd

# Specify the folder path
folder_path = r"C:\Users\GOLLA RAVI TEJA\Downloads\Property_data-20250109T055602Z-001\Property_data"

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize a list to store dataframes
dataframes = []

# Loop through the files and read them
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)  # Full file path
    try:
        # Read the file
        df = pd.read_csv(file_path, encoding='latin1')  # Specify encoding

        # Standardize column names (optional: lowercasing, removing spaces, etc.)
        df.columns = df.columns.str.lower().str.replace(' ', '_')

        # Add dataframe to the list
        dataframes.append(df)
        print(f"Read {csv_file}, shape: {df.shape}")  # Optional: print file info
    except Exception as e:
        print(f"Error reading file {csv_file}: {e}")

# Combine all dataframes into one
combined_df = pd.concat(dataframes, ignore_index=True)

# Drop invalid or irrelevant columns (example: 'unnecessary_column')
columns_to_drop = ['unnecessary_column']  # Add any invalid column names here
combined_df = combined_df.drop(columns=[col for col in columns_to_drop if col in combined_df.columns], errors='ignore')

# Display final shape
print(f"Final shape of the combined data: {combined_df.shape}")

# Display the first few rows of the combined dataframe
print(combined_df.head())
```
